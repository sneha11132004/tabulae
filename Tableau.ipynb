# Tableau

1. What is Tableau? Explain its importance in Business Intelligence and how
it helps in data-driven decision-making?
-> Tableau is a data visualization tool used to convert raw data into interactive charts, graphs, and dashboards. It helps users easily understand complex data through visuals.
In Business Intelligence, Tableau is important because it allows organizations to analyze large amounts of data quickly, identify trends, and gain useful insights. It supports real-time data analysis and improves reporting efficiency.
It helps in data-driven decision-making by providing clear visual insights, enabling users to compare data, detect patterns, and make informed business decisions instead of relying on guesswork.

2.Explain the role of the following Tableau components:
a) Data Pane
b) Worksheet
c) Dashboard
d) Story
-> a) Data Pane:
The Data Pane shows all the fields (dimensions and measures) available in the dataset. It helps users drag and drop data to create visualizations.
b) Worksheet:
A Worksheet is the area where charts and graphs are created. It is used to analyze data by building different visualizations.
c) Dashboard:
A Dashboard combines multiple worksheets into a single view. It helps in comparing different visuals and gaining insights in one place.
d) Story:
A Story is a sequence of dashboards or worksheets that present data in a narrative form. It is used to explain insights step by step.

3.What is the difference between Dimensions and Measures in Tableau?
Provide examples of each?
-> Dimensions:
Dimensions are qualitative (categorical) data used to group or label data. They are not aggregated.
Examples: Customer Name, City, Product Category, Order Date.
Measures:
Measures are quantitative (numerical) data used for calculations and analysis. They are aggregated (sum, average, etc.).
Examples: Sales, Profit, Quantity, Revenue.
Difference:
Dimensions describe data (who/what/where).
Measures calculate data (how much/how many).

4. Define and explain the purpose of Filters, Parameters, and Sets in
Tableau?
-> Filters:
Filters are used to remove unwanted data and display only relevant information in a visualization. They help users focus on specific data (e.g., showing sales for a particular region).
Parameters:
Parameters are user-defined values that allow users to change inputs dynamically. They help in creating interactive dashboards (e.g., selecting top N products).
Sets:
Sets are custom groups of data created based on conditions or selections. They help in comparing groups (e.g., top 10 customers vs others).
